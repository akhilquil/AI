```python
from sklearn.preprocessing import StandardScaler
def scaling(x):
  sclr=StandardScaler().fit(x)
  hdtf=sclr.transform(x)
  return hdtf

def postscaling(y,z):
    hdsc=pd.DataFrame(y)
    hdsc.columns=z.columns
    return hdsc
```

```python
hdsc=postscaling(scaling(dm),dm)
```

```python
hdsc
```

```python
def splitX(hdsc):
  x=pd.DataFrame()
  x=hdsc.loc[:,hdsc.columns!='SalePrice']
  return x

def splitY(hdsc):
  y=pd.DataFrame()
  y=hdsc['SalePrice']
  return y
```

```python
x=splitX(hdsc)
y=splitY(hdsc)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
def splitTrTt(x,y):
  return train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train,x_test,y_train,y_test=splitTrTt(x,y)
```

```python
def printShape():
  print("X-train shape:",x_train.shape)
  print("X-test shape:",x_test.shape)
  print("Y-train shape:",y_train.shape)
  print("Y-test shape:",y_test.shape)
printShape()
```

```python
from sklearn.linear_model import LinearRegression
def lnReg(x_train,y_train,x_test):
  ln=LinearRegression()
  ln=ln.fit(x_train,y_train)
  y_pred_train=ln.predict(x_train)
  y_pred_test=ln.predict(x_test)
  return y_pred_train,y_pred_test
```

```python
from sklearn.model_selection import GridSearchCV
def LinGridCV(x_train,y_train,x_test):
  parameters={'fit_intercept':[True,False],'normalize':[True,False],'copy_X':[True,False]}
  lr=LinearRegression()
  lr=GridSearchCV(lr,parameters,cv=15)
  lr.fit(x_train,y_train)
  lin=lr.best_estimator_
  print("Best Linear Regression Estimator on GSCV is ",lin)
  lin=lin.fit(x_train,y_train)
  y_pred_train_LinGCV=lin.predict(x_train)
  y_pred_test_LinGCV=lin.predict(x_test)
  return y_pred_train_LinGCV,y_pred_test_LinGCV
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_test_pred)
mae = mean_absolute_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R² Score:", r2)
```
