```python
from sklearn.preprocessing import StandardScaler
def scaling(x):
  sclr=StandardScaler().fit(x)
  scaled_data=sclr.transform(x)
  return scaled_data
def postscaling(y,z):
  scaled_data_df=pd.DataFrame(y)
  scaled_data_df.columns=z.columns
  return scaled_data_df
```

```python
x=dm.drop(columns='Survived')
y=dm['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y , test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)  # increase max_iter
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
accurary=accuracy_score(y_test,y_pred)
print(accurary)
```

```python
Report=classification_report(y_test,y_pred)
```
